```python
import csv
data = [['name','age','add'],
       ['Ram',34,'Kathmandu'],
       ['Shyam',56,'Bhaktapur'],
       ['Hari',78,'Lalitpur']]

file = open('data.csv','w')
x = csv.writer(file)
x.writerows(data)
file.close()
```

```python
import csv
list_data = []
file = open('data.csv')
data = csv.DictReader(file)
for i in data:
    list_data.append(i)
    
list_data
```

```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Katmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]

csvfile = open("data.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames=csv_columns)

writer.writeheader()

writer.writerows(dict)
    
csvfile.close()
```

```python
import pandas as pd
data1 = {'name':['Ram','Shyam','Hari'],
       'age':[34,56,76],
       'add':['Kathmandu','Bhaktapur','Lalitpur']}
data2 = {'name':['Ram','Shyam','Hari'],
       'age':[34,56,76],
       'add':['Kathmandu','Bhaktapur','Lalitpur']}

df1 = pd.DataFrame(data1)

df2 = pd.DataFrame(data2)

df = pd.concat([df1,df2]).reset_index()

df.to_json('concat_data.josn')
df
```
