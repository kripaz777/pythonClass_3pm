```python
# function with argument and no return type
def cal(l,b,h):                        #parameter
    a = l*b                            #local variable
    v = a*h                            #local variable
    print(a)
    print(v)
    
l = int(input("Enter l = "))       #Global variable
b = int(input("Enter b = "))       #Global variable
h = int(input("Enter h = "))       #Global variable

cal(l,b,h)                         #l,b,h argument
```

```python
def area(l,b):
    a = l*b
    print("The area = ",a)

def volume(l,b,h):
    v = l*b*h
    print("The volume = ",v)
    
l = int(input("Enter l = "))       #Global variable
b = int(input("Enter b = "))       #Global variable
h = int(input("Enter h = "))       #Global variable

area(l,b)
volume(l,b,h)
```

```python
def cal():
    l = int(input("Enter l = "))
    b = int(input("Enter b = "))
    h = int(input('Enter h = '))
    a = l*b
    v = a*h
    return a,v

x = cal()
print(x)
area,volume = x
print("The area = ",area)
print("The volume = ",volume)
```

```python
#function with argument and return type
def cal(l,b,h):
    a = l*b
    v = a*h
    return a,v

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input('Enter h = '))

x = cal(l,b,h)
area,volume = x
print("The area = ",area)
print("The volume = ",volume)
```

```python
# filter, map
```

```python
a = [1,2,3,4,5,6,7,8,9,0]
print(list(filter(lambda x : x%2==0, a)))
```

```python
a = ["apple","Ball","CAT","Dog","FISH"]
print(list(filter(lambda x: x.isupper(), a)))
```

```python
a = ["apple","Ball","CAT","Dog","FISH"]
print(list(filter(lambda x: x[0].isupper(), a)))
```

```python
# map function
a = ["Apple","Ball","CAT","Dog","FISH"]
print(list(map(lambda x: x.lower(), a)))
```

```python
a = [1,2,3,4,5,6,7,8,9]
print(list(map(lambda x: x**2, a)))
```
